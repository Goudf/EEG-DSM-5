```
import mne
import pandas as pd
import numpy as np
from scipy.fft import fft, fftfreq
from scipy.signal import butter, filtfilt
from google.colab import files
from scipy import signal
import matplotlib.pyplot as plt

#0. Ρώτησε ηλικία και φύλο ---
age = int(input("Δώσε την ηλικία: "))
gender = int(input("Δώσε φύλο (0 για άντρα, 1 για γυναίκα): "))

#1. Φόρτωσε το αρχείο EDF ---
uploaded = files.upload()
filename = list(uploaded.keys())[0]
raw = mne.io.read_raw_edf(filename, preload=True)

#2.Δημιουργία αντίγραφου για ICA
raw_ica = raw.copy()

#3.Εκτέλεση ICA για ανίχνευση artifacts
ica = mne.preprocessing.ICA(n_components=20, random_state=42)
ica.fit(raw_ica)

#4. Έλεγχος αν υπάρχουν EOG κανάλια
eog_channels_exist = any("EOG" in ch for ch in raw_ica.ch_names)

if eog_channels_exist:
    ica.find_bads_eog(raw_ica)
    ica.exclude = ica.exclude
    print("Βρέθηκαν και αφαιρέθηκαν artifacts από τα EOG κανάλια.")
else:
    print("Δεν βρέθηκαν EOG κανάλια για αφαίρεση artifacts.")

#5. Εφαρμογή ICA για να καθαριστεί το σήμα
ica.apply(raw_ica)

#6.Παίρνει δεδομένα μόνο από EEG κανάλια και απορρίπτει τα άλλα
channels = raw_ica.ch_names
eeg_channels = [ch for ch in channels if "EKG" not in ch and "EMG" not in ch and "EOG" not in ch]

data, times = raw_ica.get_data(return_times=True)
fs = int(raw.info['sfreq'])  # Συχνότητα δειγματοληψίας
df = pd.DataFrame(data.T, columns=channels)[eeg_channels]

#7. Ορισμός φίλτρου διέλευσης
lowcut = 0.5
highcut =100
lowpass_used = False
highpass_used = False

def butter_bandpass(lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    if low >= high:
        raise ValueError(f"Η τιμή του lowcut ({lowcut}) πρέπει να είναι μικρότερη από την τιμή του highcut ({highcut})")
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

#8.Βοηθητική συνάρτηση για έλεγχο FFT και επανεφαρμογή bandpass φίλτρου αν χρειάζεται
def signal_fft_with_check(signal, fs, lowcut=0.5, highcut=100):
    filtered = butter_bandpass_filter(signal, lowcut, highcut, fs)
    N = len(filtered)
    yf = np.abs(fft(filtered))[:N // 2]
    xf = fftfreq(N, 1 / fs)[:N // 2]
    if np.any((xf[yf > 0.001] < lowcut)) or np.any((xf[yf > 0.001] > highcut)):
        print("Βρέθηκαν ανεπιθύμητες συχνότητες — Επανεφαρμογή φίλτρου.")
        filtered = butter_bandpass_filter(signal, lowcut, highcut, fs)

    return filtered

#9.Εφαρμογή του φίλτρου FFT σε όλα τα κανάλια EEG
for ch in df.columns:
    df[ch] = signal_fft_with_check(df[ch], fs, lowcut=0.5, highcut=40)

#10.Υπολογισμός SNR
def compute_snr(signal, fs, signal_band=(0.5, 40), noise_band=(45, 55)):
    N = len(signal)
    yf = np.abs(fft(signal))[:N // 2]
    xf = fftfreq(N, 1 / fs)[:N // 2]

    Ps = np.sum(yf[(xf >= signal_band[0]) & (xf <= signal_band[1])] ** 2)
    Pn = np.sum(yf[(xf >= noise_band[0]) & (xf <= noise_band[1])] ** 2)

    if Pn == 0:
        return np.inf
    SNRdB = 10 * np.log10(Ps / Pn)
    return SNRdB

print("Υπολογισμός SNR πριν και μετά την επεξεργασία:\n")

for ch in eeg_channels:
    # Σήμα πριν την επεξεργασία (raw)
    raw_signal = raw.get_data(picks=[ch])[0]
    # Σήμα μετά την επεξεργασία (φιλτραρισμένο)
    filtered_signal = df[ch].values

    snr_before = compute_snr(raw_signal, fs)
    snr_after = compute_snr(filtered_signal, fs)

    print(f"Κανάλι {ch}: SNR πριν = {snr_before:.2f} dB, SNR μετά = {snr_after:.2f} dB")

#11.Εφαρμογή notch φίλτρου
Q = 30.0
def find_f0(signal, fs):
    N = len(signal)
    freqs = fftfreq(N, 1/fs)[:N//2]
    spectrum = np.abs(fft(signal))[:N//2]

    power_50 = np.sum(spectrum[(freqs > 49) & (freqs < 51)])
    power_60 = np.sum(spectrum[(freqs > 59) & (freqs < 61)])

    return 50 if power_50 > power_60 else 60

f0 =find_f0(df[df.columns[0]], fs)
print(f"Εντοπίστηκε f0 = {f0} Hz")
b_notch, a_notch = signal.iirnotch(f0, Q, fs)
for ch in df.columns:
    df[ch] = signal.filtfilt(b_notch, a_notch, df[ch])

#12.Υπολογισμός ενέργειας ανά ζώνη
def extract_band_energy(eeg_data, fs):
    bands = {
        "Delta": (0.5, 4),
        "Theta": (4, 8),
        "Alpha": (8, 13),
        "Beta": (13, 30),
        "Gamma": (35, 100)
    }
    features = {}
    for ch in eeg_data.columns:
        signal_ch = eeg_data[ch].values
        N = len(signal_ch)
        yf = np.abs(fft(signal_ch))[:N // 2]
        xf = fftfreq(N, 1 / fs)[:N // 2]
        band_energies = {}
        for band, (low, high) in bands.items():
            idx = np.where((xf >= low) & (xf <= high))
            energy = np.sum(yf[idx] ** 2)
            features[f"{ch}_{band}"] = energy
            band_energies[band] = energy

        if band_energies:
            total_energy = sum(band_energies.values())
            dominant_band = max(band_energies, key=band_energies.get)
            print(f"\n🔹 Κανάλι: {ch}")
            print(f"Κυρίαρχη Ζώνη: {dominant_band}")
            for band, energy in band_energies.items():
                pct = (energy / total_energy) * 100
                print(f"  {band}: {pct:.2f}%")
        else:
            print(f"\n Κανάλι: {ch}")
            print(" Δεν βρέθηκε κυρίαρχη ζώνη")
    return pd.DataFrame([features])

features_df = extract_band_energy(df, fs)

#13.Πρόσθεση ηλικίας και φύλου στην αρχή
features_df.insert(0, "Gender", gender)
features_df.insert(0, "Age", age)

n_channels = len(eeg_channels)

#14.Δημιουργία του plot για όλα τα κανάλια
plt.figure(figsize=(15, 2 * n_channels))

#15.Επανάληψη για κάθε κανάλι EEG
for i, channel in enumerate(eeg_channels):
    # Σήμα πριν την επεξεργασία
    raw_data, _ = raw.get_data(return_times=True)
    raw_signal = pd.DataFrame(raw_data.T, columns=channels)[channel]

    # Σήμα μετά την επεξεργασία
    filtered_signal = df[channel]

    # Δημιουργία subplot για κάθε κανάλι
    plt.subplot(n_channels, 1, i + 1)
    plt.plot(raw_signal[:1000], label="Πριν την επεξεργασία", alpha=0.7)
    plt.plot(filtered_signal[:1000], label="Μετά την επεξεργασία", alpha=0.7)
    plt.title(f"Σύγκριση σήματος EEG για το κανάλι: {channel}")
    plt.xlabel("Δείγματα")
    plt.ylabel("Απόκριση (μV)")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

#16.Αποθήκευση αποτελεσμάτων
final_df = pd.DataFrame({"Data": [",".join(features_df.astype(str).values.flatten())]})
final_df.to_excel("new_eeg_datas.xlsx", index=False)

print("Ολοκληρώθηκε! Αποθηκεύτηκαν τα χαρακτηριστικά στο 'new_eeg_datas.xlsx'")
print(features_df.head())
```
